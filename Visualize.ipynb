```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
fig = plt.figure(figsize=(18,6))
```

```python
df.shape
df.count()
```

```python
plt.title('Survived')
df.Survived.value_counts(normalize=True).plot(kind='bar', alpha=0.5)
#plt.show() ---> this shows the graph in a real python editor; in a jupyter notebook it's not necessary (i think?)
```

```python
plt.title('Age wrt Survived')
plt.scatter(df.Survived, df.Age, alpha=.1)
```

```python
plt.title('Class')
df.Pclass.value_counts(normalize=True).plot(kind='bar', alpha=0.5)
```

```python
for x in [1,2,3]:
    df.Age[df.Pclass==x].plot(kind='kde')
plt.title('Class wrt Age')
plt.legend(('1st', '2nd','3rd'))
```

```python
plt.title('Class')
df.Pclass.value_counts(normalize=True).plot(kind='kde', alpha=0.5)
```
