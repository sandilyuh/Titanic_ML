```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
fig = plt.figure(figsize=(18,6))
```

```python
df.Survived.value_counts(normalize=True).plot(kind='bar',alpha=.7)
plt.title('Survived')
```

```python
df.Survived[df.Sex == 'male'].value_counts(normalize=True).plot(kind='bar',alpha=.7)
plt.title('Men Survived')
```

```python
df.Survived[df.Sex == 'female'].value_counts(normalize=True).plot(kind='bar',alpha=.7, color='#FA0000')
plt.title('Women Survived')
```

```python
df.Sex[df.Survived == 1].value_counts(normalize=True).plot(kind='bar',alpha=.7,color='#FA0000')
plt.title('Sex Who Survived')
```

```python
df.Sex[df.Survived == 1].value_counts(normalize=True).plot(kind='bar',alpha=.7,color=['#FA0000', 'blue'])
plt.title('Sex Whom Died')
```

```python
for x in [1,2,3]:
    df.Survived[df.Pclass==x].plot(kind='kde')
plt.title('Class wrt Survived')
plt.legend(('1st', '2nd','3rd'))
```

```python
df.Survived[(df.Sex == 'male') & (df.Pclass == 1)].value_counts(normalize=True).plot(kind='bar', alpha=.7)
plt.title('1st Class Man')
```

```python
df.Survived[(df.Sex == 'male') & (df.Pclass == 2)].value_counts(normalize=True).plot(kind='bar', alpha=.7)
plt.title('2nd Class Man')
```

```python
df.Survived[(df.Sex == 'male') & (df.Pclass == 3)].value_counts(normalize=True).plot(kind='bar', alpha=.7)
plt.title('3rd Class Man')
```

```python
df.Survived[(df.Sex == 'female') & (df.Pclass == 1)].value_counts(normalize=True).plot(kind='bar', alpha=.7)
plt.title('1st Class Woman')
```

```python
df.Survived[(df.Sex == 'female') & (df.Pclass == 2)].value_counts(normalize=True).plot(kind='bar', alpha=.7)
plt.title('2nd Class Woman')
```

```python
df.Survived[(df.Sex == 'female') & (df.Pclass == 3)].value_counts(normalize=True).plot(kind='bar', alpha=.7)
plt.title('3rd Class Woman')
```
